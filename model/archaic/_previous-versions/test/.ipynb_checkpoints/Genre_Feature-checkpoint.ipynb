```python
%matplotlib inline
import pandas as pd
import pymysql
import numpy as np
import time
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Connect to the database
connection = pymysql.connect(host='localhost', user='root', password='mprophet', db='movies')

# Pull table from database
cur = connection.cursor()
    
cur.execute("select movie_id,genre FROM omdb_movie_genre a join omdb_genres b on a.genre_id=b.genre_id")
genre_data = []
for row in cur: genre_data.append(list(row))

cur.execute("select * FROM omdb_genre_list")
genre_movie_list = []
for row in cur: genre_movie_list.append(list(row))
    
cur.close()
connection.close()
```

```python
omdb_columns = ["omdb_movie_id", "omdb_genres"]

omdb_genre = pd.DataFrame(genre_data, columns=omdb_columns)

omdb_genre.head(1)
```

```python
omdb_genre['values'] = 1
omdb_table = omdb_genre.pivot(index='omdb_movie_id', columns='omdb_genres', values='values')
omdb_table = omdb_table.fillna(0)
omdb_table.head(1)
```

```python
X = np.array(omdb_table)
kmeans = KMeans(n_clusters=7, random_state=0).fit(X)
kmeans.labels_
```

```python
omdb_clusters = omdb_table.copy()
omdb_clusters['cluster'] = kmeans.labels_
omdb_clusters.head(1)
```

```python
genre_list = pd.DataFrame(genre_movie_list, columns=["omdb_movie_id","genres"])
cluster_df = pd.merge(genre_list, omdb_clusters.reset_index(),  how='inner', on=['omdb_movie_id'])

cluster_df[['omdb_movie_id','genres','cluster']].to_csv('genre.csv',index=None)
```

```python
#df.loc[df['A'] > 2, 'B'] = new_val
cluster_df_1 = cluster_df
cluster_df_1['spec']=1
cluster_df_2 = cluster_df_1[['omdb_movie_id','cluster','spec']]
```

```python
cluster_df_2.head()
```

```python
merge_cluster_df = cluster_df_2.pivot(index='omdb_movie_id', columns='cluster', values='spec')
merge_cluster_df = merge_cluster_df.fillna(0)
merge_cluster_df.head(1)
```

```python
merge_cluster_df.columns = ['genre_1','genre_2','genre_3','genre_4','genre_5','genre_6','genre_7']
```
